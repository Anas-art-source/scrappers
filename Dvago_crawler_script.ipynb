```python
# DVAGO IS THE SHOPIFY STORE

import pandas as pd
import numpy as np
import matplotlib as  plt

%matplotlib inline
pd.set_option('display.max_columns', 50)
from urllib.request import urlopen
from bs4  import BeautifulSoup

import re
import requests


baseurl  = 'https://dvago.pk/'

# header for preventing the mod_security error
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36',
}

r = requests.get(url=baseurl,headers=headers)
soup = BeautifulSoup(r.content,'lxml')
```

```python
def totalPage(link):
    parseLink = "/".join(link.split("/")[:-1])
    print(parseLink)
    r = requests.get(url=parseLink,headers=headers)
    soup = BeautifulSoup(r.content,'lxml')
    if (len(soup.select('a.pagination--item')) > 2):
        p = int(soup.select('a.pagination--item')[-2].getText().strip())
    else:
        p = 1
    return p

print(totalPage('https://dvago.pk/collections/medicine/products.json'))
```

```python
categoryMenu = [f"https://dvago.pk{a['href']}/products.json" for a in soup.select("ul.navmenu > li > a ") if a.getText().strip() == "View All"][:8]
categoryObj = [{"link": x, "page": totalPage(x)} for x in categoryMenu]
# categoryMenu = soup.select("ul.navmenu > li > a")[0].getText().strip()
categoryObj

# soup.findAll('a', class_="navmenu-link navmenu-link-parent") 
```

```python
import json
import urllib.request

def parseJson(url, pNum):
    l = f"{url}/page={pNum}"
    print(l)
    html = urllib.request.urlopen(f"{url}?page={pNum}").read()
    soup = BeautifulSoup(html,'html.parser')
    site_json=json.loads(soup.text)
    return site_json
```

```python
site_json
```

```python

def parseData(jsonObj):
    print(f"{len(jsonObj['products'])}")
    return [{"name": x["title"], "price": float(x["variants"][0]["price"]) ,"category": x["product_type"], "vendor": x["vendor"]} for x in jsonObj['products']]
```

```python
data = []

for obj in categoryObj:
    for n in range(0, obj["page"]):
         pNum = n +1
         d = parseJson(obj["link"], pNum)
         print(d)
         e = parseData(d)
         break
         data.extend(e)
```
