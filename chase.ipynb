```python
# chase crawler
import pandas as pd
import numpy as np
import matplotlib as  plt

%matplotlib inline
pd.set_option('display.max_columns', 50)
from urllib.request import urlopen
from bs4  import BeautifulSoup

import re
import requests


baseurl  = 'https://www.chasevalue.pk/'

# header for preventing the mod_security error
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36',
}
```

```python
r = requests.get(url=f"{baseurl}",headers=headers)
soup = BeautifulSoup(r.content,'lxml')
soup
```

```python
nav = soup.select('ul.site-nav > li')
categories = [{"main": x.select("a")[0]["href"] ,"sub": [y.select("li > a")[0]["href"] for y in x.select("ul")]} for x in nav]
categories
```

```python
import json
import urllib.request

def parseJson(url, pNum):
    l = f"{url}"
    print(l)
    html = urllib.request.urlopen(f"{url}?page={pNum}").read()
    soup = BeautifulSoup(html,'html.parser')
    site_json=json.loads(soup.text)
    return site_json
#     return site_json
```
