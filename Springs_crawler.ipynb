```python
# Springs IS THE SHOPIFY STORE

import pandas as pd
import numpy as np
import matplotlib as  plt

%matplotlib inline
pd.set_option('display.max_columns', 50)
from urllib.request import urlopen
from bs4  import BeautifulSoup

import re
import requests


baseurl  = 'https://springs.com.pk'

# header for preventing the mod_security error
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36',
}

r = requests.get(url=f"{baseurl}/collections",headers=headers)
soup = BeautifulSoup(r.content,'lxml')
```

```python
collections = [x['href'] for x in soup.select('div.grid-item > div.collection-item > div.img-box > a')]

collections
```

```python
import json
import urllib.request

def parseJson(url, pNum):
    l = f"{url}"
    print(l)
    html = urllib.request.urlopen(f"{url}?page={pNum}").read()
    soup = BeautifulSoup(html,'html.parser')
    site_json=json.loads(soup.text)
    return site_json
#     return site_json
```

```python
p = []
for x in collections:
    for i in range(1, 100):
        j = parseJson(f"{baseurl}{x}/products.json", i)
        products = j['products']
        print(len(products))
        if (len(products)) == 0:
            break
        productsDetail = [p.append({"category": x.split('/')[-1], "title": y["title"], "sku":y["variants"][0]["sku"]}) for y in products]
#     print(j.keys())
#     print(json)
    
    
```
